```python
import pandas as pd
from plotnine import *
from IPython.display import display
import math
from ipywidgets import VBox, Label, Button, IntSlider, Output, interact, HBox
```

# Interactive line graph

```python
@interact(w=IntSlider(value=1, min=-20, max=20), b=IntSlider(min=-20, max=20))
def g(w, b):
    data = []
    x_axis_data = list(range(-20, 20))
    for x in x_axis_data:
        data.append([x, w*x + b])
    df = pd.DataFrame(data)
    df.columns = ["X", "Y"]
    display(ggplot(df, aes(x="X", y="Y")) + geom_line() + xlim(-20, 20) + ylim(-50, 50))
    return None
```

# Line plot with different values of w and b

```python
data_dict = {
    "1.5.x + 0": {
        'w': 1.5,
        "b": 0,
        "data": None
    },
    "0.3.x + 1": {
        "w": 0.3,
        "b": 1,
        "data": None
    }
}
data = []
for d in data_dict:
    x_axis_data = list(range(-20, 20))
    for x in x_axis_data:
        data.append([x, data_dict[d]['w']*x + data_dict[d]['b'], d])
df = pd.DataFrame(data)
df.columns = ["X", "Y", "Line equation"]
plot = ggplot() + geom_line(df, aes(x="X", y="Y", color="Line equation", group="Line equation"))
display(plot)
plot.save("line_plots.png")
```

# Sigmoid plot

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
data = []
x_axis_data = list(range(-20, 20))
for x in x_axis_data:
    data.append([x, sigmoid(data_dict[d]['w']*x + data_dict[d]['b']), "\u03C3(x)"])
df = pd.DataFrame(data)
df.columns = ["X", "Y", "Line equation"]
plot = ggplot() + geom_line(df, aes(x="X", y="Y", color="Line equation", group="Line equation"))
display(plot)
plot.save("basic_sigmoid_plot.png")
```

# Sigmoid plot with different values of w and b

```python
data = []
for d in data_dict:
    x_axis_data = list(range(-20, 20))
    for x in x_axis_data:
        data.append([x, sigmoid(data_dict[d]['w']*x + data_dict[d]['b']), "\u03C3(" + d + ")"])
df = pd.DataFrame(data)
df.columns = ["X", "Y", "Line equation"]
plot = ggplot() + geom_line(df, aes(x="X", y="Y", color="Line equation", group="Line equation"))
display(plot)
plot.save("sigmoid_plots.png")
```
