```python
import pandas as pd
from plotnine import *
from IPython.display import display
import numpy as np
from ipywidgets import VBox, Label, Button, IntSlider, Output, interact, HBox
import warnings
warnings.filterwarnings('ignore')
```

# Sigmoid plot

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

# Interactive sigmoid graph

```python
X = np.array(range(-20010, 20010, 10))/1000
idea_switch_data = []
for i in X:
  if i <= 0.3:
    idea_switch_data.append([i, 1, "Ideal line equation"])
  else:
    idea_switch_data.append([i, 0, "Ideal line equation"])

@interact(w=IntSlider(value=1, min=-90, max=20), b=IntSlider(min=-30, max=30))
def g(w, b):
    data = []
    x_axis_data = np.array(range(-20010, 20010, 10))/1000
    for x in x_axis_data:
        data.append([x, sigmoid(w*x + b), "\u03C3(%s.x + %s)"%(w, b)])
    new_data = []
    new_data.extend(data)
    new_data.extend(idea_switch_data)
    df = pd.DataFrame(new_data)
    df.columns = ["X", "Y", "Line Equation"]
    loss = np.array(idea_switch_data)[:, 1].astype('float64') - np.array(data)[:, 1].astype('float64')
    display("loss : %s"%loss.mean())
    display(
        ggplot(df, aes(x="X", y="Y", color="Line Equation")) + geom_line() + ylim(-2, 2) + xlim(-2, 2) 
    )
    return None
```
